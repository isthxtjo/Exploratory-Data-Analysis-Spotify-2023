```python

# Calculate the individual correlation between bpm and musical attributes
bpmcor = data['streams'].corr(data['bpm'])
danceabilitycor = data['streams'].corr(data['danceability_%'])
energycor = data['streams'].corr(data['energy_%'])
acouscor = data['streams'].corr(data['acousticness_%'])
valencecor = data['streams'].corr(data['valence_%'])
instrucor = data['streams'].corr(data['instrumentalness_%'])
livecor = data['streams'].corr(data['liveness_%'])
speechcor = data['streams'].corr(data['speechiness_%'])
```

```python
#Group by the 'key' and calculate the average streams of the key
kstreams = data.groupby('key')['streams'].mean().sort_values()

# Group by the 'mode' and calculate the average streams for major and minor modes
mstreams = data.groupby('mode')['streams'].mean()
```

```python
# plot the keystreams and modestreams
plt.figure(figsize=(6, 3))
kstreams.plot(kind='bar', color='skyblue')
plt.xlabel('Key')
plt.ylabel('Average Streams')
plt.title('Average Streams by Key')
plt.show()

plt.figure(figsize=(3, 3))
mstreams.plot(kind='bar', color=['orange', 'purple'])
plt.xlabel('Major or Minor')
plt.ylabel('Average Streams')
plt.title('Average Streams by Mode')
plt.show()
```

```python
# Convert the other columns to numeric like what i did to streams
data['in_spotify_playlists'] = pd.to_numeric(data['in_spotify_playlists'], errors='coerce')
data['in_apple_playlists'] = pd.to_numeric(data['in_apple_playlists'], errors='coerce')
data['in_deezer_playlists'] = pd.to_numeric(data['in_deezer_playlists'], errors='coerce')
data['in_shazam_charts'] = pd.to_numeric(data['in_shazam_charts'], errors='coerce')
```

```python
#get the artist playlist count
artistcount = data.groupby('artist(s)_name')[['in_spotify_playlists', 'in_apple_playlists','in_deezer_playlists', 'in_shazam_charts']].sum()
print("Artist Playlist Counts:")
print(artistcount.sort_values(by='in_spotify_playlists', ascending=False).head(10))
```

```python
# Calculate the total playlist appearance. we'll get the result in our recent code
artistcount['Artists Total Playlist Appearances'] = artistcount.sum(axis=1)

# Sort the artist in descending order
artistcount = artistcount.sort_values(by='Artists Total Playlist Appearances', ascending=False)
print("Artist Playlist Counts (Total):")
print(artistcount[['Artists Total Playlist appearances']].head(10))
```
